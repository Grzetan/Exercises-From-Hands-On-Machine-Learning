```python
from sklearn.datasets import fetch_olivetti_faces

X, y = fetch_olivetti_faces(random_state=42, return_X_y=True)
```

```python
from sklearn.mixture import GaussianMixture
from sklearn.decomposition import PCA

pca = PCA(0.99)
mixture = GaussianMixture(n_components=50, random_state=42)
X_pca = pca.fit_transform(X)
mixture.fit(X_pca)
pca.n_components_
```

```python
import matplotlib.pyplot as plt

new_faces = mixture.sample(10)
new_faces[0].shape
```

```python
full_new_faces = pca.inverse_transform(new_faces[0])
```

```python
for face in full_new_faces:
    plt.imshow(face.reshape((64,64)), cmap='gray')
    plt.show()
```
